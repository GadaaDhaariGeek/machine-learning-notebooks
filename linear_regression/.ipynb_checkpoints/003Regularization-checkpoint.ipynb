```python
# Regularization avoids overfitting. 
# Penalizes higher coefficients, works as a shrinkage method
```

# 1. L1 Regularization

```python
# l1 regularization also does feature selection given the nature of its penalty
# it brings sparsity
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50
D = 50

# unifromly distributed numbers between 5 nd -5 
X = (np.random.random((N, D)) - .5) * 10

# true weights - only the first three features have non-zero weights
true_w = np.array([1, .5, -.5] + [0]*(D-3))

# generating Y and adding noise 
Y = X.dot(true_w) + np.random.randn(N)*.5

# performing gradient descent 
lr = .0001
max_iter = 1000
w = np.random.randn(D) / np.sqrt(D)
costs = []
l1 = 10.0 # regularization strength

for _ in range(max_iter):
    Yhat = X.dot(w)
    
    delta = Yhat - Y
    w = w - lr * (X.T.dot(delta) + l1*np.sign(w))
    
    # storing the cost 
    mse = delta.dot(delta) / N
    costs.append(mse)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(costs, marker="o")
plt.show()
```

```python
print("Final w: {}".format(w))
```

```python
plt.figure(figsize=(12, 8))
plt.plot(true_w, marker="o", label="true_w")
plt.plot(w, marker="o", label="calculated_w")
plt.legend()
plt.show()
```

# 2. L2 Regularization

```python
N = 50

# generate the data
X = np.linspace(0, 10, N)
Y = 0.5*X + np.random.normal(scale=2., size=N)

# make last two points as outliers 
Y[-1] += 30
Y[-2] += 50


plt.scatter(X, Y, edgecolor="black")
plt.show()
```

```python
# adding the bias term in X
X_b = np.hstack((np.ones(N).reshape(-1, 1), X.reshape(-1, 1)))
```

```python
# calculating the w using normal equation and plotting the values
# maximum likelihodd solution without the regularization 
w_ml = np.linalg.solve(X_b.T.dot(X_b), X_b.T.dot(Y))
Yhat_ml = X_b.dot(w_ml) 
plt.scatter(X_b[:, 1], Y, edgecolor="black")
plt.plot(X_b[:, 1], Yhat_ml, c="red", linewidth=2)
plt.show()
```

```python
# plot the regularized solution
l2 = 100.
w_map = np.linalg.solve(l2*np.eye(2) + X_b.T.dot(X_b), X_b.T.dot(Y))
Yhat_map = X_b.dot(w_map)
plt.figure(figsize=(12, 7))
plt.scatter(X_b[:, 1], Y, edgecolor="black")
plt.plot(X_b[:, 1], Yhat_ml, c="red", linewidth=2, label="without regularization")
plt.plot(X_b[:, 1], Yhat_map, c="green", linewidth=2, label="without regularization")
plt.legend()
plt.show()
```

```python
# regularization better fits and avoid overfitting than without regularization
```
