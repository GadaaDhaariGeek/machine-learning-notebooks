# 1. Gradient Descent on a general 1D curve

```python
# Gradient descent is an iterative optimization algorithm which is used to find the solution of non-convex problems
# problems for which we dont have any closed form solution is solved by gradient descent
# Although the below example is just a problem of finding minima of given function
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def J(x):
    return x*x

def g1(x):
    return 2*x

max_iter = 100
x = np.random.randint(low=10, high=100)
learning_rate = .1
loss = []


for _ in range(max_iter):
    x = x - learning_rate*g1(x)
    loss.append(J(x))
    
print("Final x: {}".format(x))
```

```python
plt.figure(figsize=(12, 8))
plt.plot(loss, marker="o")
plt.show()
```

# 2. Gradient Descent on a general 2D curve

```python
learning_rate = .01
x1, x2 = np.random.randint(10), np.random.randint(10)
max_iter = 100

def J(x1, x2):
    return (x1**2) + (x2**4)

def g1(x1):
    return 2*x1

def g2(x2):
    return 4*(x2**3)

loss = []
# Starting gradient descent 
for _ in range(max_iter):
    loss.append(J(x1, x2))
    x1 = x1 - learning_rate*g1(x1)
    x2 = x2 - learning_rate*g2(x2)

print("Final x1 = {} and x2 = {}".format(x1, x2))
```

```python
plt.figure(figsize=(12, 8))
plt.plot(loss, marker="o")
plt.show()
```

# 3. Gradient Descent for Linear Regression

```python
N = 10
D = 3
X = np.zeros((N, D))

# We can see that second column and 3rd column are collinear
# Closed form solution for this data won't work because of singularity
X[:,0] = 1 # bias term
X[:5,1] = 1
X[5:,2] = 1

Y = np.array([0]*5 + [1]*5)
```

```python
# print X so you know what it looks like
print("X:\n", X)
```

```python
costs =[]
w = np.random.randn(D) / np.sqrt(D)
learning_rate = .001
max_iter = 100
for _ in range(max_iter):
    
    # calculating Yhat
    Yhat = X.dot(w)
    # calculate delta
    delta = Yhat - Y
    # update the weights
    w = w - learning_rate * X.T.dot(delta)
    
    mse = delta.dot(delta) / N
    costs.append(mse)
```

```python
print("Final w: {}".format(w))
```

```python
plt.figure(figsize=(20, 10))
plt.plot(costs, marker="o")
plt.show()
```
