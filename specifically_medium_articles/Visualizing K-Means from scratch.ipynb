```python
import numpy as np
import matplotlib.pyplot as plt
N = 400
X1 = np.random.normal(loc=0, scale=1.35, size=(N, 2)) + np.array([0, 6])
X2 = np.random.normal(loc=0, scale=1.35, size=(N, 2)) + np.array([6, 6])
X3 = np.random.normal(loc=0, scale=1.35, size=(N, 2)) + np.array([6, 0])
X = np.vstack((X1, X2, X3))
```

```python
plt.scatter(X[:, 0], X[:, 1], edgecolor="black", c="r")
plt.show()
```
